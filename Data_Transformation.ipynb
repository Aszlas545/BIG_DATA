# 3 Wykonaj następujące obliczenia:

## 3.1. średnia liczba nowych zachorowań, śmierci i szczepień dla co najmniej 10 wybranych krajów w wybranym miesiącu

## 3.2. średnia zmiana liczby zachorowań, śmierci i szczepień dla co najmniej 10 wybranych krajów w wybranym miesiącu

## 3.3. mediana liczby nowych zachorowań, śmierci i szczepień dla co najmniej 10 wybranych krajów w wybranym miesiącu

## 3.4. odchylenie standardowe liczby nowych zachorowań, śmierci i szczepień dla co najmniej 10 wybranych krajów w wybranym miesiącu

```python
country_list = ['Poland', 'Brazil', 'France', 'Germany', 'Argentina', 'Italy', 'Sweden', 'Venezuela', 'Australia', 'United States of America']
features = ['new_confirmed', 'new_deceased', 'new_persons_vaccinated']

Means = []
Mean_diffs = []
Meds = []
Stds = []

for feature in features:

    means = []
    mean_diffs = []
    meds = []
    stds = []
    
    for country in country_list:
        query = df.query(f'date.str.contains("2021-04") and country_name == "{country}"')[feature]
        m = query.mean()
        md = query.diff().mean()
        me = query.median()
        st = query.std()
        means.append(m)
        mean_diffs.append(md)
        meds.append(me)
        stds.append(st)

    Means.append(means)
    Mean_diffs.append(mean_diffs)
    Meds.append(meds)
    Stds.append(stds)
```

```python
confirmend_df = pd.DataFrame({'country': country_list, 'mean': Means[0], 'mean difference': Mean_diffs[0], 'median': Meds[0], 'standard deviation': Stds[0]})
deacesed_df = pd.DataFrame({'country': country_list, 'mean': Means[1], 'mean difference': Mean_diffs[1], 'median': Meds[1], 'standard deviation': Stds[1]})
vaccinated_df = pd.DataFrame({'country': country_list, 'mean': Means[2], 'mean difference': Mean_diffs[2], 'median': Meds[2], 'standard deviation': Stds[2]})
```

```python
display(confirmend_df)
```

```python
plot_list = confirmend_df.drop(columns='country').columns.tolist()

figure = plt.figure(figsize=(4, 2))
figure.set_figwidth(20)
figure.set_figheight(10)
figure.subplots_adjust(top=3.4)
plt.tight_layout()
i = 0
for item in plot_list:
    fig = figure.add_subplot(4,2,i+1)
    sns.scatterplot(data=confirmend_df, x='country', y=item)
    fig.set_title(item, fontsize=15)
    plt.xticks(rotation=25)
    fig.grid()
    fig.set_ylabel('value', fontsize=10)
    i += 1

plt.show()
```

## 3.5. minimalna, średnia i maksymalna liczba nowych zachorowań, szczepień i śmierci, a PKB co najmniej 10 wybranych krajów

## 3.6. minimalna, średnia i maksymalna liczba nowych zachorowań, szczepień i śmierci, a populacja co najmniej 10 wybranych krajów.

```python
country_list = ['Poland', 'Brazil', 'France', 'Germany', 'Argentina', 'Italy', 'Sweden', 'United Kingdom', 'Australia', 'Russia']

Mins = []
Maxs = []
Means = []

for feature in features:

    mins = []
    maxs = []
    means = []
    
    for country in country_list:
        query = df.query(f'date.str.contains("2021-04") and country_name == "{country}"')[feature]
        mi = query.min()
        m = query.mean()
        ma = query.max()
        mins.append(mi)
        means.append(m)
        maxs.append(ma)

    Mins.append(mins)
    Means.append(means)
    Maxs.append(maxs)

df35 = pd.merge(pd.merge(pd.DataFrame({'country_name': country_list, 'min_new_confirmed': Mins[0], 'min_new_deceased': Mins[1], 'min_new_persons_vaccinated': Mins[2],
                     'mean_new_confirmed': Means[0], 'mean_new_deceased': Means[1], 'mean_new_persons_vaccinated': Means[2],
                     'max_new_confirmed': Maxs[0], 'max_new_deceased': Maxs[1], 'max_new_persons_vaccinated': Maxs[2]}), 
                df[['country_name', '2020 Population']].drop_duplicates().rename(columns={'2020 Population': 'Population'}), how='inner', on='country_name'),
                df2.query('Year == 2016')[['country_name', 'GDP']], how='inner', on='country_name')

with pd.option_context('display.float_format', '{:f}'.format):
    display(df35)
```

Na potrzeby tej analizy zmieniona została lista krajów ze zwględu na potrzebę połączenia dataframów, a w obu z nich były dane o róznych krajach i nie pokrywały się one.

Ponownie możemy zauważyć, że w Australii wartości znacznie odbiegają. Szczególnie dziwnym jest fakt, że bardzo mało jest danych na temat zachorowań i śmierci, a stosunkowo dużo na temat szczepień, gdzie w innych krajach trend był raczej odwrotny.

```python
plot_list = df35.drop(columns='country_name').columns.tolist()

figure = plt.figure(figsize=(6, 2))
figure.set_figwidth(20)
figure.set_figheight(15)
figure.subplots_adjust(top=3.2)
plt.tight_layout()
i = 0
for item in plot_list:
    fig = figure.add_subplot(6,2,i+1)
    sns.scatterplot(data=df35, x='country_name', y=item)
    fig.set_title(item, fontsize=15)
    plt.xticks(rotation=25)
    fig.grid()
    fig.set_ylabel('value', fontsize=10)
    fig.set_xlabel('country', fontsize=10)
    i += 1

plt.show()
```

# 4 Wykonaj normalizację dla następujących przypadków:

Przygotuj wykresy w celu porównania otrzymanych wyników. Pamiętaj o ich czytelności (typ wykresu, tytuł wykresu, podpisy osi, odpowiednie zakresy osi, itp.). Przeanalizuj otrzymane wizualizacje.

## 4.1. liczba zachorowań na określoną przez Ciebie liczbę osób (np. na 1000 osób)

```python
df['new_confirmed_per_1000'] = (df['new_confirmed'] / df['2022 Population']) * 1000
```

## 4.2. liczba szczepień na określoną przez Ciebie liczbę osób (np. na 1000 osób)

```python
df['new_vaccinated_per_1000'] = (df['new_persons_vaccinated'] / df['2022 Population']) * 1000
```

## 4.3. liczba zgonów na określoną przez Ciebie liczbę osób (np. na 1000 osób)

```python
df['new_deceased_per_1000'] = (df['new_deceased'] / df['2022 Population']) * 1000
```

## 4.4. co najmniej 4 dodatkowe, które uznasz za istotne.

```python
min_price = df['Open_MS'].min()
max_price = df['Open_MS'].max()

df['Normalized_Open_MS'] = (df['Open_MS'] - min_price) / (max_price - min_price)

min_density = df['Density (per km²)'].min()
max_density = df['Density (per km²)'].max()

df['Normalized_Density (per km²)'] = (df['Density (per km²)'] - min_density) / (max_density - min_density)

min_humidity = df['relative_humidity'].min()
max_humidity = df['relative_humidity'].max()

df['Normalized_humidity'] = (df['relative_humidity'] - min_humidity) / (max_humidity - min_humidity)

min_gold = df['Open_Gold'].min()
max_gold = df['Open_Gold'].max()

df['Normalized_Open_Gold'] = (df['Open_Gold'] - min_gold) / (max_gold - min_gold)
```

Na potrzeby zadania 4.4 użyte zostało skalowanie przy użyciu wartości minimalnych i maksymalnych.

```python
figure1 = plt.figure(figsize=(7, 3))
figure1.set_figwidth(30)
figure1.set_figheight(55)
figure1.suptitle("Porównanie nieznormalizowanych i znormalizowanych danych", fontsize=20)
figure1.subplots_adjust(top=0.95)

columns = ['new_confirmed', 'new_confirmed_per_1000', 'new_persons_vaccinated', 'new_vaccinated_per_1000', 'new_deceased', 'new_deceased_per_1000', 
           'Open_MS', 'Normalized_Open_MS', 'Density (per km²)', 'Normalized_Density (per km²)', 'relative_humidity', 'Normalized_humidity', 'Open_Gold', 'Normalized_Open_Gold']

i = 1
j = 1
for col in columns:
    if i % 2 != 0:
        col1 = col
    else:
        col2 = col
        
        fig = figure1.add_subplot(7,3,j)
        j+=1
        fig.boxplot(df[df[col1].notnull()][col1])
        fig.set_title("Nieznormalizowane " + col1, fontsize=15)
        fig.grid()
        fig.set_xticklabels([])
        plt.xlabel(col1, fontsize=10)
        fig.set_ylabel(col1, fontsize=10)
        
        fig = figure1.add_subplot(7,3,j)
        j+=1
        fig.boxplot(df[df[col2].notnull()][col2])
        fig.set_title("Znormalizowane " + col1, fontsize=15)
        fig.grid()
        fig.set_xticklabels([])
        plt.xlabel(col2, fontsize=10)
        fig.set_ylabel(col2, fontsize=10)
        
        fig = figure1.add_subplot(7,3,j)
        j+=1
        if col1 == 'Open_MS' or col1 == 'Open_Gold':
            fig.plot(df.iloc[:991][df[col1].notnull()][col1], label=col1),
            fig.plot(df.iloc[:991][df[col2].notnull()][col2], label=col2),
        else:
            fig.plot(df[df[col1].notnull()][col1], label=col1)
            fig.plot(df[df[col2].notnull()][col2], label=col2)
        plt.legend()
        plt.title("Porównanie " + col1 , fontsize=15)
        plt.grid()
        plt.xlabel('Id', fontsize=10)
        plt.ylabel("Wartość", fontsize=10)
    i += 1

plt.show()
```

Dzięki wykresom możemy odnotować na osi Oy jak zmianiła się skala wartości.

Na wykresach pudełokowych przy wartościach przeliczonych na 1000 mieszkańców możemy odnotować lepszy rozkład (outliery są znacznie bardziej rozłożone i nie nakładają się na siebie.

W przypadku wykresów liniowych linia przedstwiająca dane znormalizowane wygląda na linię prostą, jest to spowodowane tak dużą różnicą względem wartości nieznormalizowanych
